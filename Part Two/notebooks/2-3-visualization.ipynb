# Keywords

* **Matplotlib**: A comprehensive library for creating static, animated, and interactive visualizations in Python.
* **pyplot**: The matplotlib submodule that provides a MATLAB-like interface for plotting.
* **Figure**: The whole window or page that everything is drawn on.
* **Axes**: The area on which data is plotted (not to be confused with axis).
* **Axis**: The x or y axis with ticks and labels.
* **Spines**: The lines connecting the axis tick marks and noting data area boundaries.
* **Widgets**: Interactive components like sliders, buttons, and selectors.
* **Color Map (cmap)**: A mapping of data values to colors for visualization.
* **Animation**: Creating moving or updating visualizations over time.

# Basic Plotting

Let's start with the fundamental plotting functions. The `pyplot.plot()` function is the workhorse of matplotlib for creating line plots.

```python
import matplotlib.pyplot as plt
import numpy as np
# Based on matplotlib_basic_plot.py

# Generate sample points on x-axis
x = np.arange(0, 2*np.pi, 0.1)

# Create a sin function sample
y0 = np.sin(x)
plt.plot(x, y0, color='r', linewidth=3)

# Create a dashed cos function sample
y1 = np.cos(x)
plt.plot(x, y1, 'b--', linewidth=1)  # 'b--' is shorthand for blue dashed line

# Customize the axes
plt.ylim(-1, 1)
plt.xlim(0, 2*np.pi)
plt.xticks(np.arange(0, 2*np.pi, np.pi/4), 
           ['0', 'Ï€/4', 'Ï€/2', '3Ï€/4', 'Ï€', '5Ï€/4', '3Ï€/2', '7Ï€/4'])

plt.title('Sine and Cosine Functions')
plt.xlabel('Angle (radians)')
plt.ylabel('Value')
plt.legend(['sin(x)', 'cos(x)'])
plt.grid(True, alpha=0.3)
plt.show()
```

## Customizing Plot Appearance

Matplotlib provides extensive customization options:

### Color Codes
- `'r'`: Red
- `'b'`: Blue
- `'g'`: Green
- `'c'`: Cyan
- `'m'`: Magenta
- `'y'`: Yellow
- `'k'`: Black
- `'w'`: White

### Line Styles
- `'-'`: Solid line
- `'--'`: Dashed line
- `':'`: Dotted line
- `'-.'`: Dash-dot line

### Markers
- `'o'`: Circle
- `'*'`: Star
- `'+'`: Plus sign
- `'x'`: X
- `'^'`: Triangle up
- `'s'`: Square

```python
import matplotlib.pyplot as plt
import numpy as np
# Demonstrate various plot styles
x = np.linspace(0, 10, 50)

plt.figure(figsize=(10, 6))

# Different line styles and markers
plt.plot(x, np.sin(x), 'r-', linewidth=2, label='Solid red')
plt.plot(x, np.sin(x + 0.5), 'g--', linewidth=2, label='Dashed green')
plt.plot(x, np.sin(x + 1.0), 'b:', linewidth=2, label='Dotted blue')
plt.plot(x, np.sin(x + 1.5), 'c-.', linewidth=2, label='Dash-dot cyan')
plt.plot(x[::5], np.sin(x[::5] + 2.0), 'mo', markersize=8, label='Magenta circles')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Different Line Styles and Markers')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Customizing Axes

One powerful feature of matplotlib is the ability to customize axes positions and appearance. Let's explore this with the sigmoid function example.

```python
import matplotlib.pyplot as plt
import numpy as np
# Based on matplotlib_sigmoid.py

def sigmoid(x):
    """Sigmoid function: S(x) = 1/(1+e^(-x))"""
    return 1/(1 + np.exp(-x))

x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)

# Create figure with three subplots
fig = plt.figure(figsize=(15, 5))

# Subplot 1: Default axes
ax1 = fig.add_subplot(1, 3, 1)
plt.plot(x, sig, linewidth=3)
plt.title('Default Axes')
plt.grid(True, alpha=0.3)

# Subplot 2: Centered axes
ax2 = fig.add_subplot(1, 3, 2)
# Move left y-axis and bottom x-axis to centre
ax2.spines['left'].set_position('center')
ax2.spines['bottom'].set_position('center')
# Eliminate upper and right axes
ax2.spines['right'].set_color('none')
ax2.spines['top'].set_color('none')
plt.plot(x, sig, linewidth=3)
plt.title('Centered Axes')

# Subplot 3: X-axis at y=0
ax3 = fig.add_subplot(1, 3, 3)
# Move left y-axis to centre
ax3.spines['left'].set_position('center')
ax3.spines['bottom'].set_position(('data', 0))
# Eliminate upper and right axes
ax3.spines['right'].set_color('none')
ax3.spines['top'].set_color('none')
plt.plot(x, sig, linewidth=3)
plt.title('X-axis at y=0')

plt.tight_layout()
plt.show()
```

# Histogram Visualization

Histograms are useful for visualizing the distribution of data. Let's create a histogram that counts letter frequencies in text.

```python
## Example code: matplotlib_histogram.py
from matplotlib import patches
import numpy as np
import matplotlib.pyplot as plt

text = "We can know only that we know nothing. And that is the highest degree of human wisdom."
hist_labels = list('abcdefghijklmnopqrstuvwxyz') # label the 26 histogram bins with their text

data = np.zeros(0, dtype = int)
for c in text:
    if c.isalpha():
        index = int(ord(c.lower()) - ord('a'))
        data = np.append(data, index)
print(data)

max_value = np.max(data)
n, bins, patches = plt.hist(data, max_value+1, facecolor = 'green', alpha = 0.5)
plt.title("ASCII histogram")
label_axis = np.arange(0.5,26.5).tolist()
plt.xticks(label_axis, hist_labels)
plt.xlim(0, 26)
plt.show()
```

# Interactive Widgets

Matplotlib provides interactive widgets that allow users to dynamically modify visualizations. Let's explore sliders and radio buttons.

```python
# Ensure ipympl backend is installed and activated
%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, RadioButtons

# Create initial plot and values
fig, ax = plt.subplots(figsize=(10, 8))
plt.subplots_adjust(left=0.25, bottom=0.25)

t = np.arange(0.0, 1.0, 0.001)
a0 = 5  # Initial amplitude
f0 = 3  # Initial frequency
delta_f = 0.1
delta_a = 0.1

# Initial sine wave
s = a0 * np.sin(2 * np.pi * f0 * t)
l, = ax.plot(t, s, lw=2)
ax.margins(x=0)
ax.set_ylim(-10, 10)
ax.set_title('Interactive Sine Wave', fontsize=16)
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')

# Create slider axes
axcolor = 'lightgoldenrodyellow'
axfreq = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor=axcolor)
axamp = plt.axes([0.25, 0.15, 0.65, 0.03], facecolor=axcolor)

# Create sliders
sfreq = Slider(axfreq, 'Freq', 0.1, 30.0, valinit=f0, valstep=delta_f)
samp = Slider(axamp, 'Amp', 0.1, 10.0, valinit=a0, valstep=delta_a)

# Update function for sliders
def update(val):
    amp = samp.val
    freq = sfreq.val
    l.set_ydata(amp * np.sin(2 * np.pi * freq * t))
    fig.canvas.draw_idle()

sfreq.on_changed(update)
samp.on_changed(update)

# Create radio buttons for color selection
rax = plt.axes([0.025, 0.5, 0.15, 0.15], facecolor=axcolor)
radio = RadioButtons(rax, ('red', 'blue', 'green'), active=0)
l.set_color(radio.value_selected)

# Radio button update function
def colorfunc(label):
    l.set_color(label)
    fig.canvas.draw_idle()

radio.on_clicked(colorfunc)

# Explicitly display the figure
plt.show()
```

# Animation: System Clock

Matplotlib can create animations by updating plots in a loop. Let's create an analog clock that shows the current time.

```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import patches
from matplotlib.animation import FuncAnimation

%matplotlib widget

def setup_clock(ax):
    circle = patches.Circle([0, 0], 1, fill=False, edgecolor='black', linewidth=3)
    ax.add_patch(circle)

    # Draw hour markers and numbers
    for hour in range(12):
        angle = hour * 2 * np.pi / 12
        x1, y1 = 0.85 * np.cos(np.pi/2 - angle), 0.85 * np.sin(np.pi/2 - angle)
        x2, y2 = 0.95 * np.cos(np.pi/2 - angle), 0.95 * np.sin(np.pi/2 - angle)
        ax.plot([x1, x2], [y1, y2], 'k-', linewidth=2)

    for hour in range(1, 13):
        angle = hour * 2 * np.pi / 12
        x, y = 0.75 * np.cos(np.pi/2 - angle), 0.75 * np.sin(np.pi/2 - angle)
        ax.text(x, y, str(hour), ha='center', va='center', fontsize=14, fontweight='bold')

    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    ax.set_aspect('equal')
    ax.axis('off')

fig, ax = plt.subplots(figsize=(8, 8))
setup_clock(ax)

hour_hand, = ax.plot([], [], 'black', linewidth=6)
minute_hand, = ax.plot([], [], 'blue', linewidth=4)
second_hand, = ax.plot([], [], 'red', linewidth=2)
time_text = ax.set_title('', fontsize=16)

def update(frame):
    now_time = datetime.now()
    hour, minute, second = now_time.hour % 12, now_time.minute, now_time.second

    hour_angle = (hour + minute / 60) / 12 * 2 * np.pi
    minute_angle = minute / 60 * 2 * np.pi
    second_angle = second / 60 * 2 * np.pi

    def hand_coords(angle, length):
        x = length * np.cos(np.pi/2 - angle)
        y = length * np.sin(np.pi/2 - angle)
        return [0, x], [0, y]

    hour_hand.set_data(*hand_coords(hour_angle, 0.5))
    minute_hand.set_data(*hand_coords(minute_angle, 0.7))
    second_hand.set_data(*hand_coords(second_angle, 0.9))

    time_text.set_text(f"Current Time: {now_time.strftime('%H:%M:%S')}")

ani = FuncAnimation(fig, update, interval=1000)

plt.show()
```

# 3D Visualization

Matplotlib also supports 3D plotting through the `mplot3d` toolkit. Let's create a 3D spiral with scattered points.

```python
import matplotlib.pyplot as plt
import numpy as np
# Based on plot_3D.py

from mpl_toolkits import mplot3d

# Create 3D figure
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection="3d")

# Create spiral line
z_line = np.linspace(0, 15, 1000)
x_line = np.cos(z_line) * z_line
y_line = np.sin(z_line) * z_line
ax.plot3D(x_line, y_line, z_line, 'gray', linewidth=2, label='Spiral')

# Create scattered points around the spiral
n_points = 100
z_points = 15 * np.random.random(n_points)
x_points = np.cos(z_points) * z_points + 1 * np.random.randn(n_points)
y_points = np.sin(z_points) * z_points + 1 * np.random.randn(n_points)

# Scatter plot with color based on z-value
scatter = ax.scatter3D(x_points, y_points, z_points, 
                      c=z_points, cmap='hsv', s=50, alpha=0.8)

# Customize the plot
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Spiral with Random Points', fontsize=16)

# Add colorbar
plt.colorbar(scatter, ax=ax, label='Z value')

# Set viewing angle
ax.view_init(elev=20, azim=45)

plt.show()

# Create another 3D visualization - surface plot
fig2 = plt.figure(figsize=(10, 8))
ax2 = plt.axes(projection='3d')

# Create mesh grid
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Create a mathematical surface
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot surface
surf = ax2.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Add contour lines at the bottom
ax2.contour(X, Y, Z, zdir='z', offset=-1.5, cmap='viridis', levels=20)

ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_zlabel('Z axis')
ax2.set_title('3D Surface: z = sin(sqrt(xÂ² + yÂ²))', fontsize=16)
ax2.set_zlim(-1.5, 1)

# Add colorbar
plt.colorbar(surf, ax=ax2, shrink=0.5, aspect=5)

plt.show()
```

# Advanced Plotting Techniques

## Multiple Subplots and Complex Layouts

```python
import matplotlib.pyplot as plt
import numpy as np
# Create a complex figure with multiple subplot types
fig = plt.figure(figsize=(15, 10))

# Create a grid spec for complex layouts
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3, 3, figure=fig)

# Large plot on the left
ax1 = fig.add_subplot(gs[:, 0])
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), 'b-', label='sin(x)')
ax1.plot(x, np.cos(x), 'r-', label='cos(x)')
ax1.set_title('Trigonometric Functions')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Top right - scatter plot
ax2 = fig.add_subplot(gs[0, 1:])
n = 100
x_scatter = np.random.randn(n)
y_scatter = 2 * x_scatter + np.random.randn(n)
colors = np.random.rand(n)
ax2.scatter(x_scatter, y_scatter, c=colors, alpha=0.5)
ax2.set_title('Random Scatter Plot')

# Middle right - bar plot
ax3 = fig.add_subplot(gs[1, 1])
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
ax3.bar(categories, values, color=['red', 'green', 'blue', 'orange'])
ax3.set_title('Bar Chart')

# Middle far right - pie chart
ax4 = fig.add_subplot(gs[1, 2])
sizes = [30, 25, 20, 25]
ax4.pie(sizes, labels=categories, autopct='%1.1f%%', startangle=90)
ax4.set_title('Pie Chart')

# Bottom right - heatmap
ax5 = fig.add_subplot(gs[2, 1:])
data = np.random.randn(10, 10)
im = ax5.imshow(data, cmap='coolwarm', aspect='auto')
ax5.set_title('Heatmap')
plt.colorbar(im, ax=ax5)

plt.tight_layout()
plt.show()
```

# Best Practices for Data Visualization

1. **Choose the right plot type**:
   - Line plots: Time series or continuous data
   - Scatter plots: Relationships between variables
   - Bar charts: Categorical comparisons
   - Histograms: Data distributions
   - Heatmaps: 2D data or correlations

2. **Make plots readable**:
   - Use clear labels and titles
   - Choose appropriate colors
   - Add legends when necessary
   - Use appropriate font sizes

3. **Avoid chartjunk**:
   - Remove unnecessary decorations
   - Use minimal grid lines
   - Focus on the data

4. **Consider your audience**:
   - Technical vs. general audience
   - Color blindness considerations
   - Cultural color associations
